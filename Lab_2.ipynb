```python
# Exercise 1
def func1(a, b):
  result = 0
  for i in range(0, b):
    result += a
  return result
```

```python
# Sample function call: func1(5, 3)
import time
st=time.time()
print(func1(4,3))
end=time.time()
print("Running time in ms:",(end-st)*1000)
#the function performs multiplication using repeated addition
```

```python
# Exercise 2
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
  return result
```

```python
# Sample function call: func2(2, 4)
st=time.time()
print(func2(2,4))#This function func2(a,b) returns repeted multiplication of a until b is positive
end=time.time()
print("Running time in ms:",(end-st)*1000)
```

```python
# Exercise 3
def func3(num):
   result = 0
   while num > 0:
    result = result + num % 10
    num //= 10
    return result
```

```python
# Sample function call: func3(1234)
st=time.time()
ans=func3(1234)
print(ans)# this function counts number of digits
end=time.time()
print("Running time in ms:",(end-st)*1000)
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
st=time.time()
print(func4([1, 2, 3, 4], [2, 4, 6, 8])) # this function checks if 2 array has common element
end=time.time()
print("Running time in ms:",(end-st)*1000)
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
st=time.time()
print(func5(36)) # this function returns the perfect sq if not ,none
end=time.time()

print((end-st)*1000)
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)
    return A

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - i + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
    elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)

```

```python
#Sample function call: func8([3,1,2,7,8,6])
l=[3,1,2,7,8,6]
st=time.time()

print(func8(l)) # the functon recursively breaks array in 3 subparts and sort them seperately
end=time.time()
print((end-st)*1000)
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
st=time.time()
print(func12([1, 2, 3, 4, 5], 6)) #The function returns a new array by shifting all elememts by k steps
end=time.time()
print((end-st)*1000)
```

```python
# write a function named as sum_arr to find sum of elements in given array

def sum_arr(arr):
  sum=0
  for i in arr:
    sum+=i
  return sum
```

```python
#call the function
arr = [12, 3, 4, 15,45,78,450]
st=time.time()
print(sum_arr(arr))
end=time.time()
print((end-st)*1000)
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array

def Counting_EvenOdd(arr):
  oc=0
  ec=0
  for i in arr:
    if i%2==0:
      ec+=1
    else:
      oc+=1
  print("Odd elements count:",oc)
  print("Even elements count:",ec)
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr,tar):
   for i in range(0,len(arr)):
    if arr[i]==tar:
      return i
   return -1
```

```python
# call the function liner_Search for array = [1, 3, 5, 7, 9]
array = [1, 3, 5, 7, 9]
st=time.time()
print(linear_Search(array,9))
end=time.time()
print((end-st)*1000)
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr,tar):
  i=0
  j=len(arr)-1

  while(i<=j):
    mid=i+(j-i)//2

    if arr[mid]==tar:
       return mid
    elif arr[mid]>tar:
      j=mid-1
    else:
     i=mid+1
  return -1
```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
arr = [1, 3, 5, 7, 9,15,16,14,45]
st=time.time()
print(Binary_Search(arr,45))
end=time.time()
print((end-st)*1000)
```
